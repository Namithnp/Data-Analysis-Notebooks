```python
def nbc(data, inp, label):
    l = []
    for values in inp.values():
        l.append(values)
    #Set up individual classes with appended input row
    class_labels = np.unique(data[label])
    class_grouped = data.groupby(by = data[label])

    print("Class 1: ", class_labels[0])
    class1 = class_grouped.get_group(class_labels[0])

    print("Class 2: ", class_labels[1])
    class2 = class_grouped.get_group(class_labels[1])

    class1_prob = (len(class1))/len(data)
    class2_prob = (len(class2))/len(data)

    #Probability calculation
    print("Calculating class probability for ", class_labels[0])
    class1 = class1.drop(columns=label)
    prob1 = 1
    for i, col in enumerate(class1):
        feat, counts = np.unique(class1[col], return_counts=True)
        res = dict(zip(feat, counts))
        if l[i] in res:
            prob1 = prob1 * res[l[i]]/len(class1)
        else:
            prob1=0
    final_prob1 = prob1 * class1_prob
    print(final_prob1)

    print("Calculating class probability for ", class_labels[1])
    class2 = class2.drop(columns=label)
    prob2 = 1
    for i, col in enumerate(class2):
        feat, counts = np.unique(class2[col], return_counts=True)
        res = dict(zip(feat, counts))
        if l[i] in res:
            prob2 = prob2 * res[l[i]]/len(class2)
        else:
            prob2=0
    final_prob2 = prob2 * class2_prob
    print(final_prob2)


    if final_prob1 > final_prob2:
        print(f'{class_labels[0]} is the outcome')
    else:
        print(f'{class_labels[1]} is the outcome')

```
